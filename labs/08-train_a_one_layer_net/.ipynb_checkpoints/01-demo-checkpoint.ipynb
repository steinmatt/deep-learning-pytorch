#           TRAINING A ONE LAYER NET ON MNIST

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from random import randint
import utils
```

### randint(a,b) returns a random integer between a and b: 

```python
idx=randint(5,10)

print(idx)
```

### Download the TRAINING SET (data+labels)

```python
train_data=torch.load('../../data/mnist/train_data.pt')

print(train_data.size())
```

```python
train_label=torch.load('../../data/mnist/train_label.pt')

print(train_label.size())
```

### Download the TEST SET (data only)

```python
test_data=torch.load('../../data/mnist/test_data.pt')

print(test_data.size())
```

### Make a one layer net class

```python
class one_layer_net(nn.Module):

    def __init__(self, input_size, output_size):
        super().__init__()
        self.linear_layer = nn.Linear( input_size, output_size , bias=False)
        
    def forward(self, x):
        x = self.linear_layer(x)
        p = F.softmax(x, dim=1)
        return p
```

### Build the net

```python
net = one_layer_net(784,10)
print(net)
```

### Take the 6th image of the test set:

```python
im = test_data[6]
utils.show(im)
```

### And feed it to the UNTRAINED network:

```python
p =  net( im.view(1,784) ) 
print(p)
```

### Display visually the confidence scores

```python
utils.show_prob_mnist(p)
```

### Train the network (only 5000 iterations) on the train set

```python
criterion = nn.NLLLoss()
optimizer = torch.optim.SGD(net.parameters() , lr=0.01 )

for iter in range(1,5000):
    
    # choose a random integer between 0 and 59,999 
    # extract the corresponding picture and label
    idx = randint(0, 60000-1)
    im = train_data[idx]
    label = train_label[idx]
    
    # reshape the image to fit the network
    input = im.view(1,784)

    # feed the input to the net  
    prob = net(input) 
 
    # update the weights (all the magic happens here -- we will discuss it later)
    log_prob = torch.log(prob)
    loss = criterion(log_prob, label.view(1))    
    optimizer.zero_grad()       
    loss.backward()
    optimizer.step()
```

### Take the 6th image of the test set:

```python
im=test_data[6]
utils.show(im)
```

### Feed it to the TRAINED net:

```python
prob =  net( im.view(1,784)) 
print(prob)
```

### Display visually the confidence scores

```python
utils.show_prob_mnist(prob)
```
