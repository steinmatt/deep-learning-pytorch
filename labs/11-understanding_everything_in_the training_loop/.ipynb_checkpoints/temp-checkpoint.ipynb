# Understanding Everything in the training loop -- DEMO

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from random import randint
import utils
```

### Download the data and print the sizes

```python
train_data=torch.load('../../data/mnist/data/train_data.pt')

print(train_data.size())
```

```python
train_label=torch.load('../../data/mnist/data/train_label.pt')

print(train_label.size())
```

```python
test_data=torch.load('../../data/mnist/data/test_data.pt')

print(test_data.size())
```

### Make a three layer net class

```python
class three_layer_net(nn.Module):

    def __init__(self, input_size, hidden_size1, hidden_size2,  output_size):
        super(three_layer_net , self).__init__()
        
        self.layer1 = nn.Linear( input_size,   hidden_size1 , bias=False)
        self.layer2 = nn.Linear( hidden_size1, hidden_size2 , bias=False)
        self.layer3 = nn.Linear( hidden_size2, output_size  , bias=False)
        
    def forward(self, x):
        
        y       = self.layer1(x)
        y_hat   = F.relu(y)
        z       = self.layer2(y_hat)
        z_hat   = F.relu(z)
        scores  = self.layer3(z_hat)
        
        return scores
    
    
    
net=three_layer_net(784, 1000, 1500, 10)  


```

### Build the net

```python
net=three_layer_net(784, 1000, 1500, 10)
print(net)
```

### Choose the criterion and optimizer

```python

criterion = nn.CrossEntropyLoss()


optimizer=torch.optim.SGD( net.parameters() , lr=0.01 )

```

### Choose the size of the mini-batches 

```python
bs=200
```

### Train the network (only 5000 iterations) on the train set

```python
for iter in range(1,5000):
    
    # Set dL/dU, dL/dV, dL/dW to be filled with zeros
    optimizer.zero_grad()
     
    # create a minibatch
    indices=torch.LongTensor(bs).random_(0,60000)
    minibatch_data =  train_data[indices]
    minibatch_label= train_label[indices]
    
    #reshape the minibatch
    inputs = minibatch_data.view(bs,784)
    
    # tell Pytorch to start tracking all operations that will be done on it
    inputs.requires_grad_()

    # forward the minibatch through the net  
    scores=net( inputs ) 
    
    # Compute the average of the losses of the data points in the minibatch
    loss = criterion( scores , minibatch_label) 
    
    # backward pass to compute dL/dU, dL/dV and dL/dW    
    loss.backward()
    
    # do one step of stochastic gradient descent: U=U-lr(dL/dU), V=V-lr(dL/dU), ...
    optimizer.step()
    
    print(loss.detach())
    
```

```python
for iter in range(5000):
    
    optimizer.zero_grad()
     
    indices = torch.LongTensor(bs).random_(0,60000)
    minibatch_data  =  train_data[indices]
    minibatch_label = train_label[indices]
    
    inputs = minibatch_data.view(bs,784)
    
    inputs.requires_grad_()

    scores = net( inputs ) 
    
    loss = criterion( scores , minibatch_label ) 
        
    loss.backward()
    
    optimizer.step()
    
```
